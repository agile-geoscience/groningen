# Read formation tops

- Read tops to dictionaries
- Read tops to `striplog`
- Read tops to `pandas`

## Raw data

```python
from striplog.utils import read_petrel
import numpy as np

fname = "../data/Formation_tops/Well_Tops.asc"

# Need striplog 0.8.8 to pass a single function.
nullify = lambda x: np.nan if x==-999 else x

data = read_petrel(fname, function=nullify)
data.keys()
```

## To `pandas`

```python
import pandas as pd

df = pd.DataFrame.from_dict(data)
df.head()
```
